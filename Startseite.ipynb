1. Wählen Sie ein Themengebiet aus der oben angezeigten Liste aus, sodass sich die entsprechende Datei in einem neuen Reiter öffnet.
2. Betätigen Sie innerhalb des entsprechenden Reiters die Schaltfläche mit dem Doppelpfeil _"Restart the Kernel, then re-run the whole notebook"_. Diese befindet sich unterhalb der Reiterübersicht. <br>
